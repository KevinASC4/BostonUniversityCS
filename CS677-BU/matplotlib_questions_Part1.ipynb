### Part One: Creating data from an equation

It is important to be able to directly translate a real equation into a plot. Your first task actually is pure numpy, then we will explore how to plot it out with Matplotlib. The [world famous equation](https://en.wikipedia.org/wiki/Mass%E2%80%93energy_equivalence) from Einstein:

$$E=mc^2$$


Use your knowledge of Numpy to create two arrays: E and m, where **m** is simply 11 evenly spaced values representing 0 grams to 10 grams. E should be the equivalent energy for the mass. You will need to figure out what to provide for **c** for the units m/s, a quick google search will easily give you the answer (let's use the close approximation in our solutions).

```python
# CODE HERE
import numpy as np
```

```python
m = np.arange(11)
print(m)
```

```python
# CODE HERE
c = 3.0e8  
E = m * c**2
print(E)
```

### Part Two: Plotting $E=mc^2$

Now that we have the arrays E and m, we can plot this to see the relationship between Energy and Mass. 

Import what you need from Matplotlib to plot out graphs.

**TASK: Recreate the plot shown below which maps out $E=mc^2$ using the arrays we created in the previous task. Note the labels, titles, color, and axis limits.**

You should attempt to re-create each major component. Better recreations will receive higher points.

Use the <tt>figure</tt> and <tt>axis</tt> objects to complete the task.

```python
# CODE HERE
import numpy as np
import matplotlib.pyplot as plt

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(8, 5))

# Plot the data
ax.plot(m, E, linestyle='-', color='r',)

# Set labels and title
ax.set_xlabel('Mass in grams')
ax.set_ylabel('Energy in Joules')
ax.set_title('E=Mc^2')

# Set axis limits (assuming we match the expected plot)
ax.set_xlim(0, 10)
ax.set_ylim(0, max(E) * 1.1)  # Adding a bit of space above max value

# Add grid and legend
ax.grid(True)
ax.legend()

# Show the plot
plt.show()

```
