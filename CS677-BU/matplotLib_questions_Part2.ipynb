1. We will continue with our plotting of the $E=mc^2$ equation from the previous homework. Can you figure out how to plot the *mass vs Energy* plot on a *logarthimic* scale on the y axis? Place a grid along the y axis ticks as well. The plot shown here only required minor modifications of the code for the changes.
 - You will have to carefully review the options associated with the `plt.grid()` command for this question.  

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
c = 3.0e8 

# Create mass array (0 to 10 grams, converted to kg)
m = np.linspace(0, 10, 11) 

# Compute energy using E = mc^2
E = m * c**2

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(8, 5))

# Plot the data
ax.plot(m, E, linestyle='-', color='r')

# Set y-axis to logarithmic scale
ax.set_yscale('log')

# Set labels and title
ax.set_xlabel('Energy in Joules')
ax.set_title('Mass vs Energy on a Logarithmic Scale')

# Add grid along the y-axis ticks only
ax.grid(True, which='both', axis='y')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

3. This problem will test your ability to generate subplots of varying sizes and types.
   1. Data: generate a 1D array `x` comprising 100 equally spaced points between 0 and 10.
   2. Using  `x` as input, generate y-axis data corresponding to the following three functions: $sin()$, $e^{()}$, and $\log(\cdot+1)$
   3. Use the `subplot2grid()` method to reproduce the plot below as closely as possible. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate the x array
x = np.linspace(0, 10, 100)

y_sin = np.sin(x)
y_exp = np.exp(-x)  # Exponential decay
y_log = np.log(x + 1)

# Create a figure
fig = plt.figure(figsize=(12, 8))

# Subplot for sin(x) - occupies the top left
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax1.plot(x, y_sin, color='b', label='sin(x)')  # No lines or markers
ax1.set_title('Sine Wave')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.legend()


# Subplot for e^(-x) - occupies the bottom left
ax2 = plt.subplot2grid((2, 2), (1, 0))
ax2.plot(x, y_exp, color='g', label='e^(-x)')  # No lines or markers
ax2.set_title('Exponential Decay')
ax2.set_xlabel('x')
ax2.set_ylabel('e^(-x)')


# Subplot for log(x + 1) - occupies the right half
ax3 = plt.subplot2grid((2, 2), (0, 1), rowspan=2)
ax3.plot(x, y_log, color='r', label='log(x + 1)')  # No lines or markers
ax3.set_title('Logarithmic Curve')
ax3.set_xlabel('x')
ax3.set_ylabel('log(x + 1)')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
